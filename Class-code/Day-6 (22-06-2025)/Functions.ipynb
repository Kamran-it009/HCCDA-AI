```python
def sum1(a, b): # a and b are parameters.
    c = a + b
    return c # returned value
```

```python
def sum2(a = 20, b = 15): # Default parameters.
    x = a + b
    return x # returned value
```

```python
def sum3(a, b): # a and b are parameters.
    print(a)
    print(b)
    c = a + b
    return c # returned value
```

```python
sum3(b=3, a=4)
```

```python
sum4 = lambda a, b: a + b
```

```python
sum4(34, 5)
```

```python
def sum_all(*a):
    sum = 0  # 10, 89, 10, 89, 78, 14, 15
    for i in a:
        sum = i + sum
    
    return sum
```

```python
sum_all(2, 3, 4, 5, 6, 6, 7, 90, 10, 78, 60)
```
