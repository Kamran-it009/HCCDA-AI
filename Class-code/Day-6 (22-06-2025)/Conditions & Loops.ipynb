```python
fruits = ['apple', 'banana', 'cherry', 'mango', 'orange', 'almod', 'melon']
```

```python
for i in fruits:
    print(f'{i}-Sweet')
```

```python
fruits_dict = {'A': 'apple', 'B': 'banana', 
               'C': 'cherry', 'D': 'Dragon fruits'}
```

```python
for i in fruits_dict:
    print(i)
```

```python
for i in fruits_dict.values():
    print(i)
```

```python
for i in fruits_dict.items():
    print(i)
```

```python
for i, j in fruits_dict.items():
    print(f'{i} - {j}')
```

```python
for i in range(20):
    if i%2 == 0:
        print(i)
```

```python
x = [i for i in range(20) if i % 2 == 0]
print(x)
```

```python
for n in range(10):
    if n == 6:
        break
    print(f'{n}-Kamran')
```
